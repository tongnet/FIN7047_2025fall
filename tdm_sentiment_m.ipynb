```python
import pandas as pd
import os

# Set working directory
os.chdir(r"d:\tong\teaching\fin7047\2025fall\project")

# Load the CSV (first row is header by default)
df = pd.read_csv(r"d:\tong\teaching\fin7047\2025fall\project\btcsent_tdm.csv", header=0)

df
```

```python
# Convert Date column to datetime
df[date_col] = pd.to_datetime(df[date_col])

# ✅ Convert sentiment column to numeric (coerce errors -> NaN)
df[score_col] = pd.to_numeric(df[score_col], errors='coerce')

# Extract year and month
df['YearMonth'] = df[date_col].dt.to_period('M')

# Group by month and calculate average score
monthly_avg = df.groupby('YearMonth')[score_col].mean().reset_index()

# Convert YearMonth back to string for saving
monthly_avg['YearMonth'] = monthly_avg['YearMonth'].astype(str)

# Save result to new CSV
output_path = r"d:\tong\teaching\fin7047\2025fall\project\monthly_average_score.csv"
monthly_avg.to_csv(output_path, index=False)

print("✅ Monthly average score saved to:", output_path)
print(monthly_avg)
```
